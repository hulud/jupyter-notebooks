```python
from __future__ import division
import numpy
from numpy import genfromtxt
from scipy.signal import savgol_filter
import  matplotlib.pyplot as plt
import peakutils

plt.xkcd(scale=2, randomness=2)

IonizationSignal_name = 'Intensity '
IonizationSignal_units = 'arb. units'
wavenumber_name = 'Wavenumber'
wavenumber_units = 'cm$^{-1}$'
axisFormat = '{} ({})'
xAxisLabel = axisFormat.format(wavenumber_name,wavenumber_units)
yAxisLabel = axisFormat.format(IonizationSignal_name,IonizationSignal_units)


fig = plt.figure()
ax = plt.gca()
#ax.tick_params(axis='y',left='off')



spectrumFiles = ['110418112703UV_572.5-574.5z=-0.655.DAT',
                 '110418142329UV_572.5-574.5z=-0.655.DAT']
          #       '160418133435UV_572.5-574.5z=-0.858.DAT']

wlCol = 0
wnCol = 1
timeCol = 2
ionSigCol = 3

spectra = [genfromtxt(spectrumFile, skip_header=1) for spectrumFile in spectrumFiles]

avgIonSignal = numpy.mean(numpy.array([spectrum[:,ionSigCol] for spectrum in spectra]), axis=0)
filteredIonizationSignal = savgol_filter(avgIonSignal, 5, 2)
baseLineSignal = peakutils.baseline(filteredIonizationSignal)
correctedIonizationSignal = filteredIonizationSignal - baseLineSignal

wn = spectra[0][:, wnCol]
peakIndexes = peakutils.indexes(correctedIonizationSignal, thres=0.1, min_dist=15)
enhancedPeaks_wn = peakutils.interpolate(wn, correctedIonizationSignal, ind=peakIndexes)


#plt.plot(wn, avgIonSignal)
plt.plot(wn,correctedIonizationSignal)
#print enhancedPeak_wn
peakNames = ['A', 'B', 'D', 'C$_1$/C$_2$', 'E','F','']
for peakIndex, peakName, enhancedPeak_wn in zip(peakIndexes, peakNames, enhancedPeaks_wn):
    ax.text(enhancedPeak_wn, correctedIonizationSignal[peakIndex], peakName, fontsize=12,  ha='left', va='bottom')

plt.title('Whoo Hoo!!!')
#plt.scatter(enhancedPeak_wn,correctedIonizationSignal[peakIndexes])
#print avgIonSig
plt.tight_layout()

plt.savefig('Tryptamine_R2PI')
plt.show()
```
